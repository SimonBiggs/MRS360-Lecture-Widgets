```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0, 1)
y = np.linspace(0, 2)
xx, yy = np.meshgrid(x, y)

z_radians = np.exp(-xx) * np.sin(yy)

plt.figure(figsize=(7,7))
contours_radians = plt.contour(x, y, z_radians, 20, cmap='viridis')
plt.plot(-np.log(0.5/np.sin(1.68)), 1.68, 'kx', ms=10, mew=3)
plt.plot(-np.log(0.5/np.sin(2)), 2, 'kx', ms=10, mew=3)
plt.clabel(contours_radians)

plt.title(r"Contour plot of $z(x,y) = e^{-x}\times sin(y)$ with y in radians")
plt.xlabel("x")
plt.ylabel("y (in radians)")
```

```python
z_degrees = np.exp(-xx) * np.sin(yy * np.pi / 180)

plt.figure(figsize=(7,7))
contours_degrees = plt.contour(x, y, z_degrees, 20, cmap='viridis')
plt.clabel(contours_degrees)

plt.title(r"Contour plot of $z(x,y) = e^{-x}\times sin(y)$ with y in degrees")
plt.xlabel("x")
plt.ylabel("y (in degrees)")
```
